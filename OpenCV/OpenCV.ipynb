This script demonstrates how to concatenate (join) two images using NumPy and display them with OpenCV. First, it reads the same image (healthy3.jpg) twice using cv2.imread() and stores them in img1 and img2. Then, it horizontally stacks the two images side by side using np.hstack() and displays the result in a window titled "Concatenated Image" for 1 second. After closing the window, it vertically stacks the same two images one on top of the other using np.vstack() and again displays the concatenated result in the same window title. Finally, it closes the window using cv2.destroyAllWindows(). This method is useful for visual comparison or montage creation, provided both images have the same dimensions and number of channels.

```python
import numpy as np
img1= cv2.imread('healthy3.jpg')
img2= cv2.imread('healthy3.jpg')

img=np.hstack((img1 , img2))
# Display the concatenated image
cv2.imshow('Concatenated Image', img)
cv2.waitKey(1000)
cv2.destroyAllWindows()

img=np.vstack((img1 , img2))
cv2.imshow('Concatenated Image', img)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

This Python script uses OpenCV to display an image with overlaid text. The image 'healthy3.jpg' is read using cv2.imread(). Then, cv2.putText() is called to draw the word 'Healthy' on the image at position (50, 50) using the FONT_HERSHEY_DUPLEX font. The bottomLeftOrigin=False flag ensures that the origin for the text position starts from the top-left corner of the image.
```python
img = cv2.imread('healthy3.jpg')

txt = cv2.putText(
        img=img,                         # The image on which to draw
        text='Healthy',                  # The string/text to be drawn
        org=(50, 50),                    # Bottom-left corner of the text in the image
        fontFace=cv2.FONT_HERSHEY_DUPLEX,  # Font type
        fontScale=1,                     # Font scale factor that is multiplied by the font-specific base size
        color=(0, 255, 0),               # Text color in BGR (Green here)
        thickness=2,                     # Thickness of the lines used to draw the text
        lineType=cv2.LINE_AA,            # Type of the line (Anti-aliased)
        bottomLeftOrigin=False           # When true, the origin is at the bottom-left corner. Otherwise, it's top-left.
    )
cv2.imshow('Text on Image', txt)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

This script loads two images, resizes them to the same dimensions (640x480), and performs various pixel-wise operations to compare and combine them. It first subtracts the second image from the first using cv2.subtract, highlighting the differences between the two. Then, it applies bitwise operations: bitwise_and to get the intersection (common pixels), bitwise_or to combine all pixels from both images, and bitwise_xor to highlight the pixels that are different in either image but not both. Each resulting image is displayed in a separate window for 2 seconds, illustrating how these operations affect the images visually.

```python
img1= cv2.imread('healthy3.jpg')
img2= cv2.imread('healthy leaf.jpg')

img1=cv2.resize(img1,(640, 480))
img2=cv2.resize(img2,(640, 480))

new_img1 = cv2.subtract(
    src1=img1, 
    src2=img2  # Subtract the second image from the first
)
new_img2 = cv2.bitwise_and(
    src1=img1, 
    src2=img2  # Subtract the second image from the first
)
new_img3 = cv2.bitwise_or(
    src1=img1, 
    src2=img2  # Subtract the second image from the first
)
new_img4 = cv2.bitwise_xor(
    src1=img1, 
    src2=img2  # Subtract the second image from the first
)
cv2.imshow('Subtracted Image', new_img1)
cv2.waitKey(2000)
cv2.imshow('Bitwise AND Image', new_img2)
cv2.waitKey(2000)
cv2.imshow('Bitwise OR Image', new_img3)
cv2.waitKey(2000)
cv2.imshow('Bitwise XOR Image', new_img4)
cv2.waitKey(2000)

cv2.destroyAllWindows()
```

This Python script uses OpenCV to load, resize, rotate, and display an image. It begins by loading the image 'healthy3.jpg' using cv2.imread() and then resizes it to 640×480 pixels with cv2.resize(), applying linear interpolation for smooth scaling. After obtaining the width and height of the resized image, the script calculates a rotation matrix using cv2.getRotationMatrix2D() to rotate the image 45 degrees around its center without changing its scale. The rotation is applied using cv2.warpAffine(), which transforms the image based on the rotation matrix. Finally, the rotated image is displayed in a window for 2 seconds using cv2.imshow() and cv2.waitKey(), after which the window is closed with cv2.destroyAllWindows(). This process demonstrates basic image transformation techniques in OpenCV.

```python
img = cv2.imread('healthy3.jpg')
res_img= cv2.resize(
    src=img, 
    dsize=(640, 480),  # Desired size of the output image
    interpolation=cv2.INTER_LINEAR  # Interpolation method for resizing
)

w,h = res_img.shape[1], res_img.shape[0]

m = cv2.getRotationMatrix2D((w/2 , h/2) ,45,0.5)
rotated_img = cv2.warpAffine(res_img , m , (w,h))
cv2.imshow('Rotated Image', rotated_img)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

This Python script demonstrates three different types of image blurring techniques using OpenCV and visually compares them side by side.

First, the image 'healthy3.jpg' is loaded with cv2.imread() and resized to 200×200 pixels for uniform processing. Then, three kinds of blurring filters are applied:

Gaussian Blur (cv2.GaussianBlur()): This filter smooths the image by averaging pixel values using a Gaussian kernel of size (5,5). It is useful for reducing image noise and detail.

Median Blur (cv2.medianBlur()): This replaces each pixel's value with the median of surrounding pixels in a 5x5 kernel. It is particularly effective in removing salt-and-pepper noise.

Bilateral Filter (cv2.bilateralFilter()): This filter preserves edges while smoothing the image. It uses three parameters: diameter of pixel neighborhood (9), and two values controlling how much the filter considers color and space (both set to 75 here).

The original and blurred images are horizontally stacked using np.hstack() for side-by-side comparison, and displayed in a window using cv2.imshow(). The window stays open for 2 seconds (cv2.waitKey(2000)) before being closed with cv2.destroyAllWindows(). This script helps visualize the differences between various blurring techniques in OpenCV.

```python
img = cv2.imread('healthy3.jpg')

img = cv2.resize(img, (200, 200))

gussian_blur = cv2.GaussianBlur(img, (5,5) , 0)
median_blur = cv2.medianBlur(img, 5)
bileteral_blur = cv2.bilateralFilter(img, 9, 75, 75)

h_Stack = np.hstack((img, gussian_blur, median_blur, bileteral_blur))
cv2.imshow('Blurred Images', h_Stack)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

In this code, an image named 'healthy3.jpg' is first read and resized to a 200×200 pixel format for uniformity and better visualization. A 5×5 kernel (matrix of ones) is created using NumPy, which acts as the structuring element for all the morphological operations. Then, four key transformations are applied. First, erosion (cv2.erode) is used, which shrinks the white regions in the image and removes small white noise by applying the kernel, effectively thinning the objects. Second, dilation (cv2.dilate) is performed, which does the opposite of erosion — it enlarges the white regions, helping to fill in small holes or gaps. Next, the top-hat transformation (cv2.MORPH_TOPHAT) is computed, which extracts small bright elements on a dark background by subtracting the result of an opening operation from the original image. Lastly, a morphological gradient (cv2.MORPH_GRADIENT) is applied, which highlights the edges of objects by computing the difference between the dilated and eroded versions of the image. All the resulting images, including the original, are horizontally stacked using np.hstack for side-by-side comparison and displayed using cv2.imshow. The window stays open for 2 seconds (cv2.waitKey(2000)) before all OpenCV windows are closed using cv2.destroyAllWindows.

Erosion (cv2.erode)

What it does: Shrinks the white regions (foreground) in an image.

Use-case: Removes small white noises, disconnects two connected objects.

Dilation (cv2.dilate)

What it does: Expands the white regions.

Use-case: Useful for joining broken parts of an object or increasing object area.

Opening (cv2.morphologyEx(..., MORPH_OPEN, ...))

What it does: Erosion followed by dilation.

Use-case: Removes small objects (noise) from the foreground.

Closing (cv2.morphologyEx(..., MORPH_CLOSE, ...))

What it does: Dilation followed by erosion.

Use-case: Fills small holes inside the foreground objects.

Morphological Gradient (cv2.morphologyEx(..., MORPH_GRADIENT, ...))

What it does: Difference between dilation and erosion.

Use-case: Highlights the edges of objects.

Top Hat (cv2.morphologyEx(..., MORPH_TOPHAT, ...))

What it does: Difference between the input image and its opening.

Use-case: Detects bright spots on dark backgrounds.

Black Hat (cv2.morphologyEx(..., MORPH_BLACKHAT, ...))

What it does: Difference between the closing and the input image.

Use-case: Detects dark spots on bright backgrounds.

```python
import numpy as np
import cv2
img = cv2.imread('healthy3.jpg')
img =cv2.resize(img, (200, 200))

m = np.ones((5 ,5), np.int64)

en = cv2.erode(
    src=img, 
    kernel=m,  # Structuring element for erosion
    iterations=1  # Number of times erosion is applied
)
di = cv2.dilate(
    src=img,
    kernel=m,  # Structuring element for dilation
    iterations=1  # Number of times dilation is applied
)
tp = cv2.morphologyEx(
    src=img, 
    op=cv2.MORPH_TOPHAT,  # Top-hat transformation
    kernel=m  # Structuring element for the operation
)

gr = cv2.morphologyEx(
    src=img, 
    op=cv2.MORPH_GRADIENT,  # Morphological gradient
    kernel=m  # Structuring element for the operation
)
h_stack = np.hstack((img, en, di , tp, gr))
cv2.imshow('Erosion and Dilation', h_stack)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

This code performs background subtraction on a video to isolate moving objects (foreground) from a static background. It begins by loading a video file (testVedio.mp4) using OpenCV's VideoCapture. A background subtractor object is created using cv2.createBackgroundSubtractorMOG2(), which is a common algorithm for background subtraction. The code reads the video frame by frame in a loop. Each frame is resized to 600×500 pixels for uniform display. The background subtractor is then applied to the current frame to produce a foreground mask, which highlights moving elements while suppressing static background areas. Both the original resized frame and the resulting foreground mask are displayed in separate windows. The loop continues until the video ends or the user presses the 'q' key. Finally, the video capture object is released, and all OpenCV windows are closed.

```python
cap = cv2.VideoCapture(r'C:\Users\SHAYANTAN BISWAS\Desktop\Jupyter\OpenCV\Vedio\testVedio.mp4')
sub_bg = cv2.createBackgroundSubtractorMOG2()  # Create a background subtractor object

while cap.isOpened():
    ret, frame = cap.read()  # Read a frame from the video
    if ret==True:
        frame = cv2.resize(frame, (600, 500))  # Resize the frame to 600x500
        fg_mask = sub_bg.apply(frame)  # Apply the background subtractor to get the foreground mask
        cv2.imshow('Foreground Mask', fg_mask)
        cv2.imshow('Original Frame', frame)
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break
    else:
        break    
    
cap.release()  # Release the video capture object    
cv2.destroyAllWindows()  # Release the camera and close all OpenCV windows    
```

cv2.warpPerspective(img, m, (w, h)) applies this perspective transformation to the image, warping it so the original 400x400 image is "squeezed" into the smaller 200x200 area but output on a 400x400 canvas.

```python
import cv2
import numpy as np

def nothing(x):
    pass

# Create a window
cv2.namedWindow("Trackbars")

# Create trackbars for lower HSV
cv2.createTrackbar("LH", "Trackbars", 0, 179, nothing)
cv2.createTrackbar("LS", "Trackbars", 0, 255, nothing)
cv2.createTrackbar("LV", "Trackbars", 0, 255, nothing)

# Create trackbars for upper HSV
cv2.createTrackbar("UH", "Trackbars", 179, 179, nothing)
cv2.createTrackbar("US", "Trackbars", 255, 255, nothing)
cv2.createTrackbar("UV", "Trackbars", 255, 255, nothing)

# Start video capture
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to HSV color space
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Get current positions of all trackbars
    lh = cv2.getTrackbarPos("LH", "Trackbars")
    ls = cv2.getTrackbarPos("LS", "Trackbars")
    lv = cv2.getTrackbarPos("LV", "Trackbars")

    uh = cv2.getTrackbarPos("UH", "Trackbars")
    us = cv2.getTrackbarPos("US", "Trackbars")
    uv = cv2.getTrackbarPos("UV", "Trackbars")

    # Define lower and upper HSV bounds
    lower_bound = np.array([lh, ls, lv])
    upper_bound = np.array([uh, us, uv])

    # Create mask
    mask = cv2.inRange(hsv, lower_bound, upper_bound)

    # Optional: Morphological filtering to clean up mask
    kernel = np.ones((5, 5), np.uint8)
    mask = cv2.erode(mask, kernel, iterations=1)
    mask = cv2.dilate(mask, kernel, iterations=2)

    # Find contours
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on original frame
    output = frame.copy()
    cv2.drawContours(output, contours, -1, (0, 255, 0), 2)

    # Show frames
    cv2.imshow("Original", frame)
    cv2.imshow("Mask", mask)
    cv2.imshow("Contours", output)

    # Press ESC to exit
    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture('a-man-running-side-view-animation-reference-ytshorts.savetube.me.mp4')  # Open the default camera (0 for the first camera, 1 for the second, etc.)

while cap.isOpened():
    ret, frame = cap.read()
    
    if ret==True:
     
        frame = cv2.resize(frame, (800, 500))
   
        frame = cv2.flip(frame, 1)
        
        gry = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        har_cascade = cv2.CascadeClassifier('haarcascade_fullbody.xml')  # Load the Haar Cascade classifier for full body detection
        
        body = har_cascade.detectMultiScale(
            image=gry,  # Source image
            scaleFactor=1.1,  # Parameter specifying how much the image size is reduced at each image scale
            minNeighbors=5,  # Parameter specifying how many neighbors each candidate rectangle should have to retain it
            minSize=(30, 30)  # Minimum possible object size
        )
        for (x, y, w, h) in body:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
            
        cv2.imshow('Original Frame', frame)
        if cv2.waitKey(20) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()  # Release the video capture object
cv2.destroyAllWindows()  # Release the camera and close all OpenCV windows
        
```
